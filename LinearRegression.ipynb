```python
from sklearn.datasets import fetch_california_housing
```

```python
housing = fetch_california_housing()
```

```python
print(housing.feature_names)
```

```python
# Prepearing the dataset 
import pandas as pd
dataset = pd.DataFrame(housing.data, columns=housing.feature_names)
```

```python
print(housing.target)
```

```python
dataset['price']=housing.target
```

```python
# figure out the correlation (multicolinearity)
import seaborn as sns
```

```python
from matplotlib import style
style.use("dark_background")

sns.pairplot(dataset)
```

```python
dataset
```

```python
print(housing.DESCR)

import matplotlib.pyplot as plt
for column in dataset.columns[:-1]:  # Exclude the target column
    plt.figure(figsize=(6, 4))  # Set figure size
    sns.scatterplot(x=dataset[column], y=dataset["price"])
    plt.title(f"{column} vs Price")  # Title for clarity
    plt.xlabel(column)
    plt.ylabel("Price (MedHouseVal)")
    plt.show()
```

```python
for column in dataset.columns[:-1]:  # Exclude the target column
    plt.figure(figsize=(6, 4))  # Set figure size
    sns.regplot(x=dataset[column], y=dataset["price"])
    plt.title(f"{column} vs Price")  # Title for clarity
    plt.xlabel(column)
    plt.ylabel("Price (MedHouseVal)")
    plt.show()
```

```python
# price is my dependent feature and remaining all are dependent features
X=dataset.iloc[:,:-1]
Y=dataset.iloc[:,-1]
```

```python
X.head()
```

```python
Y.head()
```

```python
# Train test split 
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=42)
```

```python
X_train
```

```python
X_test
```

```python
Y_train
```

```python
Y_test
```

```python
# standard scaling 
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
X_train=scaler.fit_transform(X_train)
```

```python
X_test=scaler.transform(X_test)
```

```python
X_train
```

```python
X_test
```

```python
# Model training 
from sklearn.linear_model import LinearRegression
```

```python
regression=LinearRegression()
```

```python
regression.fit(X_train, Y_train)
```

```python
# print regression and coefficients 
print(regression.coef_)
```

```python
print(regression.intercept_)
```

```python
# on which params the regression has been trained 
regression.get_params()
```

```python
# prediction with test data 
reg_pred=regression.predict(X_test)
reg_pred
```

```python
# plot a scatter plot 
plt.scatter(Y_test, reg_pred) # not linear 
```

```python
residuals=Y_test-reg_pred
sns.displot(residuals, kind="kde")
# normal distribution -> model doing good 
```

```python
# scatter plot with respect to predicion and residuals
plt.scatter(reg_pred, residuals)
```

```python
import numpy as np
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error

print(mean_absolute_error(Y_test, reg_pred))
print(mean_squared_error(Y_test, reg_pred))
print(np.sqrt(mean_squared_error(Y_test, reg_pred)))
```

```python
# adjusted r2 Performace metrics 
from sklearn.metrics import r2_score
score=r2_score(Y_test, reg_pred)
score
```

```python
# adjusted r2
1 - (1 - score) * (len(Y_test) - 1) / (len(Y_test) - X_test.shape[1] - 1)
```
